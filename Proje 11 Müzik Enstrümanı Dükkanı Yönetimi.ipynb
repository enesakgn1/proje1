```python
import tkinter as tk

class Enstruman:
    def __init__(self, ad, stok_miktari):
        self.ad = ad
        self.stok_miktari = stok_miktari

    def satis_yap(self, satilan_adet):
        if self.stok_miktari >= satilan_adet:
            self.stok_miktari -= satilan_adet
            return True
        else:
            print("Stokta yeterli miktarda enstruman yok!")
            return False

    def destek_olustur(self, talep_detaylari):
        print(f"Destek talebi oluşturuldu: {talep_detaylari}")

class Siparis:
    def __init__(self, musteri, enstruman, adet):
        self.musteri = musteri
        self.enstruman = enstruman
        self.adet = adet

class Musteri:
    def __init__(self, ad):
        self.ad = ad
        self.siparisler = []

musteri1 = Musteri("Ahmet")
musteri2 = Musteri("Mehmet")
musteriler = [musteri1, musteri2]

enstrumani = [Enstruman("Gitar", 10), Enstruman("Keman", 5), Enstruman("Davul", 3)]

def enstrumanları_goster():
    for i, enstruman in enumerate(enstrumani):
        tk.Label(pencere, text=f"{i+1}. {enstruman.ad}: {enstruman.stok_miktari} adet stokta").pack()

def enstruman_sat():
    global secili_musteri
    if secili_musteri is not None and secili_musteri < len(musteriler):
        satilan_enstruman_ad = enstruman_adi.get()
        satilan_adet = int(adet.get())
        for enstruman in enstrumani:
            if enstruman.ad == satilan_enstruman_ad:
                if enstruman.satis_yap(satilan_adet):
                    siparis = Siparis(musteriler[secili_musteri].ad, satilan_enstruman_ad, satilan_adet)
                    musteriler[secili_musteri].siparisler.append(siparis)
                    print(f"{musteriler[secili_musteri].ad} {satilan_adet} {satilan_enstruman_ad} aldı.")
                    break
    else:
        print("Lütfen bir müşteri seçin.")

def destek_talebi_olustur(ozellik):
    global secili_musteri
    if secili_musteri is not None and secili_musteri < len(musteriler):
        talep_detaylari = ozellik
        for enstruman in enstrumani:
            enstruman.destek_olustur(talep_detaylari)
    else:
        print("Lütfen bir müşteri seçin.")

def secili_musteri_degisti(event):
    global secili_musteri
    secili_musteri = musteri_listbox.curselection()[0] if musteri_listbox.curselection() else None

pencere = tk.Tk()
pencere.title("Müzik Enstrümanı Dükkanı Yönetimi")

enstruman_adi = tk.StringVar()
adet = tk.IntVar()

tk.Label(pencere, text="Müşteri seçin:").pack()

musteri_listbox = tk.Listbox(pencere, height=10)
for musteri in musteriler:
    musteri_listbox.insert(tk.END, musteri.ad)
musteri_listbox.pack()

musteri_listbox.bind("<<ListboxSelect>>", secili_musteri_degisti)

secili_musteri = None

tk.Label(pencere, text="Enstrüman adı:").pack()
tk.Entry(pencere, textvariable=enstruman_adi).pack()

tk.Label(pencere, text="Adet:").pack()
tk.Entry(pencere, textvariable=adet).pack()

tk.Button(pencere, text="Enstrüman sat", command=enstruman_sat).pack()

tk.Label(pencere, text="Destek talebi detayı:").pack()

destek_talebi_secenekler = ["İade yapmak istiyorum", "Kargom nerede", "Başka bir konuda destek talebi"]
secim = tk.StringVar()
secim.set(destek_talebi_secenekler[0]) 

dropdown_menu = tk.OptionMenu(pencere, secim, *destek_talebi_secenekler)
dropdown_menu.pack()

def talep_yarat():
    destek_talebi_olustur(secim.get())

tk.Button(pencere, text="Destek talebi oluştur", command=talep_yarat).pack()

enstrumanları_goster()

pencere.mainloop()
```
