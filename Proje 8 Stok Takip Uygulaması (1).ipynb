```python
class StokTakip:
    def __init__(self):
        self.stok = {}

    def urun_ekle(self, urun_adı, adet):
        if urun_adı in self.stok:
            self.stok[urun_adı] += adet
        else:
            self.stok[urun_adı] = adet

    def urun_cikar(self, urun_adı, adet):
        if urun_adı in self.stok:
            if self.stok[urun_adı] >= adet:
                self.stok[urun_adı] -= adet
            else:
                print("Stokta yeterli miktarda ürün bulunmamaktadır.")
        else:
            print("Belirtilen ürün stokta bulunmamaktadır.")

    def stok_listele(self):
        for urun, miktar in self.stok.items():
            print(f"{urun}: {miktar} adet")
```

```python
import tkinter as tk
from tkinter import messagebox

class StokTakipGUI:
    def __init__(self, root, stok_takip):
        self.root = root
        self.root.title("Stok Takip Sistemi")
        self.stok_takip = stok_takip

        self.urun_listesi = tk.Listbox(root, width=50)
        self.urun_listesi.grid(row=0, column=0, columnspan=2)

        self.urun_ekle_label = tk.Label(root, text="Ürün Adı:")
        self.urun_ekle_label.grid(row=1, column=0)
        self.urun_ekle_entry = tk.Entry(root)
        self.urun_ekle_entry.grid(row=1, column=1)

        self.adet_label = tk.Label(root, text="Adet:")
        self.adet_label.grid(row=2, column=0)
        self.adet_entry = tk.Entry(root)
        self.adet_entry.grid(row=2, column=1)

        self.urun_ekle_button = tk.Button(root, text="Ürün Ekle", command=self.urun_ekle)
        self.urun_ekle_button.grid(row=3, column=0, columnspan=2)

        self.urun_cikar_label = tk.Label(root, text="Ürün Adı:")
        self.urun_cikar_label.grid(row=4, column=0)
        self.urun_cikar_entry = tk.Entry(root)
        self.urun_cikar_entry.grid(row=4, column=1)

        self.adet_cikar_label = tk.Label(root, text="Adet:")
        self.adet_cikar_label.grid(row=5, column=0)
        self.adet_cikar_entry = tk.Entry(root)
        self.adet_cikar_entry.grid(row=5, column=1)

        self.urun_cikar_button = tk.Button(root, text="Ürün Çıkar", command=self.urun_cikar)
        self.urun_cikar_button.grid(row=6, column=0, columnspan=2)

    def urun_ekle(self):
        urun_adı = self.urun_ekle_entry.get()
        adet = self.adet_entry.get()
        if urun_adı and adet.isdigit():
            adet = int(adet)
            self.stok_takip.urun_ekle(urun_adı, adet)
            self.urun_listesi.delete(0, tk.END)
            for urun, miktar in self.stok_takip.stok.items():
                self.urun_listesi.insert(tk.END, f"{urun}: {miktar} adet")
            messagebox.showinfo("Başarılı", f"{urun_adı} başarıyla eklendi.")
        else:
            messagebox.showerror("Hata", "Ürün adı boş olamaz ve adet sayısı geçerli bir sayı olmalıdır.")

    def urun_cikar(self):
        urun_adı = self.urun_cikar_entry.get()
        adet = self.adet_cikar_entry.get()
        if urun_adı and adet.isdigit():
            adet = int(adet)
            self.stok_takip.urun_cikar(urun_adı, adet)
            self.urun_listesi.delete(0, tk.END)
            for urun, miktar in self.stok_takip.stok.items():
                self.urun_listesi.insert(tk.END, f"{urun}: {miktar} adet")
            messagebox.showinfo("Başarılı", f"{urun_adı} başarıyla çıkarıldı.")
        else:
            messagebox.showerror("Hata", "Ürün adı boş olamaz ve adet sayısı geçerli bir sayı olmalıdır.")

class StokTakip:
    def __init__(self):
        self.stok = {}

    def urun_ekle(self, urun_adı, adet):
        if urun_adı in self.stok:
            self.stok[urun_adı] += adet
        else:
            self.stok[urun_adı] = adet

    def urun_cikar(self, urun_adı, adet):
        if urun_adı in self.stok:
            if self.stok[urun_adı] >= adet:
                self.stok[urun_adı] -= adet
            else:
                print("Stokta yeterli miktarda ürün bulunmamaktadır.")
        else:
            print("Belirtilen ürün stokta bulunmamaktadır.")

if __name__ == "__main__":
    root = tk.Tk()
    stok_takip = StokTakip()
    app = StokTakipGUI(root, stok_takip)
    root.mainloop()
```
