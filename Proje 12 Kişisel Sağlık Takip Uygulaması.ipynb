```python
class Kullanıcı:
    def __init__(self, adı, yaş, cinsiyet):
        self.adı = adı
        self.yaş = yaş
        self.cinsiyet = cinsiyet
        self.sağlık_kayıtları = []

    def kayıt_ekle(self, sağlık_kaydı):
        self.sağlık_kayıtları.append(sağlık_kaydı)

    def egzersiz_ekle(self, egzersiz):
        pass

    def rapor_oluştur(self):
        pass

class SağlıkKaydı:
    def __init__(self, kullanıcı, tarih, kan_basıncı, nabız, kilo, boy):
        self.kullanıcı = kullanıcı
        self.tarih = tarih
        self.kan_basıncı = kan_basıncı
        self.nabız = nabız
        self.kilo = kilo
        self.boy = boy

class Egzersiz:
    def __init__(self, adı, süresi, tekrarları):
        self.adı = adı
        self.süresi = süresi
        self.tekrarları = tekrarları
```

```python
import tkinter as tk

class HealthApp:
    def __init__(self, ana_pencere):
        self.ana_pencere = ana_pencere
        ana_pencere.title("Kişisel Sağlık Takip Uygulaması")

        self.etiket = tk.Label(ana_pencere, text="Hoş geldiniz!")
        self.etiket.pack()

        self.saglik_kaydet_button = tk.Button(ana_pencere, text="Sağlık Verisi Kaydet", command=self.saglik_verisi_kaydet)
        self.saglik_kaydet_button.pack()

        self.egzersiz_ekle_button = tk.Button(ana_pencere, text="Egzersiz Ekle", command=self.egzersiz_ekle)
        self.egzersiz_ekle_button.pack()

        self.rapor_olustur_button = tk.Button(ana_pencere, text="Rapor Oluştur", command=self.rapor_olustur)
        self.rapor_olustur_button.pack()

    def saglik_verisi_kaydet(self):
        saglik_penceresi = tk.Toplevel(self.ana_pencere)
        saglik_penceresi.title("Sağlık Verisi Kaydet")

        tk.Label(saglik_penceresi, text="Tarih:").pack()
        self.tarih_entry = tk.Entry(saglik_penceresi)
        self.tarih_entry.pack()

        tk.Label(saglik_penceresi, text="Kan Basıncı (Sistolik/Diyastolik):").pack()
        self.kan_entry = tk.Entry(saglik_penceresi)
        self.kan_entry.pack()

        tk.Label(saglik_penceresi, text="Nabız:").pack()
        self.nabiz_entry = tk.Entry(saglik_penceresi)
        self.nabiz_entry.pack()

        tk.Label(saglik_penceresi, text="Kilo (kg):").pack()
        self.kilo_entry = tk.Entry(saglik_penceresi)
        self.kilo_entry.pack()

        tk.Label(saglik_penceresi, text="Boy (cm):").pack()
        self.boy_entry = tk.Entry(saglik_penceresi)
        self.boy_entry.pack()

        kaydet_button = tk.Button(saglik_penceresi, text="Kaydet", command=self.saglik_verisi_kaydet_kaydet)
        kaydet_button.pack()

        kapat_button = tk.Button(saglik_penceresi, text="Kapat", command=saglik_penceresi.destroy)
        kapat_button.pack()

    def saglik_verisi_kaydet_kaydet(self):
        tarih = self.tarih_entry.get()
        kan = self.kan_entry.get()
        nabiz = self.nabiz_entry.get()
        kilo = self.kilo_entry.get()
        boy = self.boy_entry.get()

        print("Sağlık verileri kaydedildi:")
        print("Tarih:", tarih)
        print("Kan Basıncı:", kan)
        print("Nabız:", nabiz)
        print("Kilo:", kilo)
        print("Boy:", boy)
        
    def egzersiz_ekle(self):
        egzersiz_penceresi = tk.Toplevel(self.ana_pencere)
        egzersiz_penceresi.title("Egzersiz Ekle")

        tk.Label(egzersiz_penceresi, text="Egzersiz Adı:").pack()
        tk.Entry(egzersiz_penceresi).pack()

        tk.Label(egzersiz_penceresi, text="Süre (dk):").pack()
        tk.Entry(egzersiz_penceresi).pack()

        tk.Label(egzersiz_penceresi, text="Tekrarlar:").pack()
        tk.Entry(egzersiz_penceresi).pack()

        kaydet_button = tk.Button(egzersiz_penceresi, text="Kaydet", command=self.egzersiz_ekle_kaydet)
        kaydet_button.pack()

        kapat_button = tk.Button(egzersiz_penceresi, text="Kapat", command=egzersiz_penceresi.destroy)
        kapat_button.pack()

    def egzersiz_ekle_kaydet(self):
        print("Egzersiz kaydedildi.")
        
    def rapor_olustur(self):
        rapor_penceresi = tk.Toplevel(self.ana_pencere)
        rapor_penceresi.title("Rapor Oluştur")

        tk.Label(rapor_penceresi, text="Rapor:").pack()
        self.rapor_text = tk.Text(rapor_penceresi, height=10, width=40)
        self.rapor_text.pack()

        kaydet_button = tk.Button(rapor_penceresi, text="Kaydet", command=self.rapor_olustur_kaydet)
        kaydet_button.pack()

        kapat_button = tk.Button(rapor_penceresi, text="Kapat", command=rapor_penceresi.destroy)
        kapat_button.pack()

    def rapor_olustur_kaydet(self):
        rapor = self.rapor_text.get("1.0", tk.END)
        print("Rapor kaydedildi:")
        print(rapor)

root = tk.Tk()
app = HealthApp(root)
root.mainloop()
```
