```python
import tkinter as tk

class Ders:
    def __init__(self, adi, konular, ogretmen):
        self.adi = adi
        self.konular = konular
        self.ogretmen = ogretmen

class Ogrenci:
    def __init__(self, adi, soyadi, sinif):
        self.adi = adi
        self.soyadi = soyadi
        self.sinif = sinif

class EgitimMateryaliUygulamasi:
    def __init__(self, root):
        self.root = root
        self.root.title("Eğitim Konu Paylaşım Platformu")

        self.ogrenciler = [
            Ogrenci("Ahmet", "Yılmaz", "10-A"),
            Ogrenci("Ayşe", "Demir", "10-A"),
            Ogrenci("Mehmet", "Kaya", "10-A"),
            Ogrenci("Fatma", "Arslan", "10-A"),
            Ogrenci("Emre", "Çakrak", "10-A"),
            Ogrenci("Zeynep", "Koç", "10-A"),
            Ogrenci("Emre", "Yıldırım", "10-A"),
            Ogrenci("Selin", "Çelik", "10-A"),
            Ogrenci("Burak", "Aksoy", "10-A"),
            Ogrenci("Gizem", "Turan", "10-A")
        ]

        self.dersler = [
            Ders("Matematik", ["polinomlar", "Olasılık", "Denklemler"], "Ahmet Hoca"),
            Ders("Fizik", ["Vektörler", "Atışlar", "Dalgalar"], "Ayşe Hoca"),
            Ders("Kimya", ["Elementler", "Periyodik Tablo", "Karışımlar"], "Fatma Hoca"),
            Ders("Tarih", ["Beylikten Devlete", "Dünya Gücü Osmanlı", "Osmanlının Çöküşü"], "Efe Hoca"),
            Ders("Coğrafya", ["Beşeri Sistemler", "Katmanlar", "Nüfüs"], "Zeynep Hoca")
        ]

        self.ogrenci_label = tk.Label(root, text="Öğrenci Seçiniz:")
        self.ogrenci_label.pack(pady=10)

        self.ogrenci_secim = tk.StringVar(root)
        self.ogrenci_secim.set("Öğrenci Seçiniz")
        self.ogrenci_menu = tk.OptionMenu(root, self.ogrenci_secim, *[f"{ogrenci.adi} {ogrenci.soyadi} - {ogrenci.sinif}" for ogrenci in self.ogrenciler])
        self.ogrenci_menu.pack(pady=5)

        self.ders_label = tk.Label(root, text="Ders Seçiniz:")
        self.ders_label.pack(pady=10)

        self.ders_secim = tk.StringVar(root)
        self.ders_secim.set("Ders Seçiniz")
        self.ders_menu = tk.OptionMenu(root, self.ders_secim, *[ders.adi for ders in self.dersler])
        self.ders_menu.pack(pady=5)

        self.konu_label = tk.Label(root, text="Konular:")
        self.konu_label.pack(pady=10)

        self.konu_liste = tk.Listbox(root, width=50)
        self.konu_liste.pack(pady=5)

        self.soru_label = tk.Label(root, text="Soru Sorun:")
        self.soru_label.pack(pady=10)

        self.soru_entry = tk.Entry(root, width=50)
        self.soru_entry.pack(pady=5)

        self.soru_button = tk.Button(root, text="Soru Sor", command=self.soru_sor)
        self.soru_button.pack(pady=5)

        self.ogrenci_secim.trace("w", self.dersleri_guncelle)
        self.ders_secim.trace("w", self.konulari_guncelle)

    def dersleri_guncelle(self, *args):
        self.ders_menu['menu'].delete(0, 'end')
        for ders in self.dersler:
            self.ders_menu['menu'].add_command(label=ders.adi, command=tk._setit(self.ders_secim, ders.adi))

    def konulari_guncelle(self, *args):
        secilen_ders = next((ders for ders in self.dersler if ders.adi == self.ders_secim.get()), None)
        if secilen_ders:
            self.konu_liste.delete(0, tk.END)
            for konu in secilen_ders.konular:
                self.konu_liste.insert(tk.END, konu)

    def soru_sor(self):
        secilen_ogrenci = self.ogrenci_secim.get()
        secilen_ders = self.ders_secim.get()
        secilen_konu_index = self.konu_liste.curselection()
        secilen_konu = self.konu_liste.get(secilen_konu_index)
        soru = self.soru_entry.get()

        print("Seçilen Öğrenci:", secilen_ogrenci)
        print("Seçilen Ders:", secilen_ders)
        print("Seçilen Konu:", secilen_konu)
        print("Soru:", soru)

root = tk.Tk()
uygulama = EgitimMateryaliUygulamasi(root)
root.mainloop()
```
