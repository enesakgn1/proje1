```python
import tkinter as tk
from tkinter import messagebox

class Hasta:
    def __init__(self, isim, tc, randevu_gecmisi=None):
        self.isim = isim
        self.tc = tc
        self.randevu_gecmisi = randevu_gecmisi if randevu_gecmisi else []

    def randevu_al(self, randevu):
        self.randevu_gecmisi.append(randevu)

    def randevu_iptal(self, randevu):
        self.randevu_gecmisi.remove(randevu)


class Doktor:
    def __init__(self, isim, uzmanlik_alani, musaitlik_durumu=True):
        self.isim = isim
        self.uzmanlik_alani = uzmanlik_alani
        self.musaitlik_durumu = musaitlik_durumu

    def __str__(self):
        return f"{self.isim} ({self.uzmanlik_alani})"


class Randevu:
    def __init__(self, tarih, doktor, hasta):
        self.tarih = tarih
        self.doktor = doktor
        self.hasta = hasta


class RandevuSistemiGUI:
    def __init__(self, root):
        self.root = root
        self.root.title("Hasta Randevu Sistemi")

        self.doktorlar = [
            Doktor("Dr. Ahmet", "Dahiliye"),
            Doktor("Dr. Ayşe", "Kulak Burun Boğaz"),
            Doktor("Dr. Mehmet", "Göz Hastalıkları")
        ]

        self.hastalar = [
            Hasta("Ali", "13245678904"),
            Hasta("Sude", "23456789012"),
            Hasta("Emre", "423855678908"),
            Hasta("Efe", "62345678902"),
            Hasta("Berat", "76345688906")
        ]

        self.randevu_label = tk.Label(root, text="Randevu Tarihi:")
        self.randevu_label.grid(row=0, column=0)
        self.randevu_entry = tk.Entry(root)
        self.randevu_entry.grid(row=0, column=1)

        self.doktor_label = tk.Label(root, text="Doktor Seçimi:")
        self.doktor_label.grid(row=1, column=0)
        self.doktor_var = tk.StringVar(root)
        self.doktor_var.set(str(self.doktorlar[0]))
        self.doktor_optionmenu = tk.OptionMenu(root, self.doktor_var, *[str(d) for d in self.doktorlar])
        self.doktor_optionmenu.grid(row=1, column=1)

        self.hasta_label = tk.Label(root, text="Hasta TC:")
        self.hasta_label.grid(row=2, column=0)
        self.hasta_var = tk.StringVar(root)
        self.hasta_var.set(self.hastalar[0].tc)
        self.hasta_optionmenu = tk.OptionMenu(root, self.hasta_var, *[h.tc for h in self.hastalar])
        self.hasta_optionmenu.grid(row=2, column=1)

        self.randevu_al_button = tk.Button(root, text="Randevu Al", command=self.randevu_al)
        self.randevu_al_button.grid(row=3, column=0, columnspan=2)

        self.randevu_bilgileri_button = tk.Button(root, text="Randevu Bilgileri", command=self.randevu_bilgileri_goster)
        self.randevu_bilgileri_button.grid(row=4, column=0, columnspan=2)

    def randevu_al(self):
        randevu_tarihi = self.randevu_entry.get()
        randevu_tarihi = self.format_tarih(randevu_tarihi)  # Tarih formatını düzenle
        doktor_secimi = self.doktor_var.get()
        hasta_tc = self.hasta_var.get()

        for doktor in self.doktorlar:
            if str(doktor) == doktor_secimi:
                secilen_doktor = doktor
                break

        for hasta in self.hastalar:
            if hasta.tc == hasta_tc:
                secilen_hasta = hasta
                break

        randevu = Randevu(randevu_tarihi, secilen_doktor, secilen_hasta)
        secilen_hasta.randevu_al(randevu)
        messagebox.showinfo("Bilgi", "Randevu başarıyla alındı.")

    def format_tarih(self, tarih):
        # Tarih formatını düzenle
        if len(tarih) == 8 and tarih.isdigit():
            return f"{tarih[:2]}.{tarih[2:4]}.{tarih[4:]}"
        return tarih

    def randevu_bilgileri_goster(self):
        randevular = ""
        for hasta in self.hastalar:
            if hasta.randevu_gecmisi:
                randevular += f"\n{hasta.isim} randevu geçmişi:\n"
                for randevu in hasta.randevu_gecmisi:
                    randevular += f"Tarih: {randevu.tarih}, Doktor: {randevu.doktor.isim}\n"
        if randevular:
            messagebox.showinfo("Randevu Bilgileri", randevular)
        else:
            messagebox.showinfo("Randevu Bilgileri", "Herhangi bir randevu bulunamadı.")


root = tk.Tk()
app = RandevuSistemiGUI(root)
root.mainloop()
```
