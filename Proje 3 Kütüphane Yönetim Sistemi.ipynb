```python
class Kitap:
    def __init__(self, kitap_id, ad, yazar, durum=True):
        self.kitap_id = kitap_id
        self.ad = ad
        self.yazar = yazar
        self.durum = durum

    def durum_guncelle(self, durum):
        self.durum = durum


class Uye:
    def __init__(self, ad, soyad, uye_id):
        self.ad = ad
        self.soyad = soyad
        self.uye_id = uye_id


class Odunc:
    def __init__(self):
        self.odunc_listesi = []

    def odunc_al(self, kitap, uye, tarih):
        if kitap.durum:
            kitap.durum_guncelle(False)
            self.odunc_listesi.append((kitap, uye, tarih))
        else:
            print("Kitap zaten ödünçte.")

    def iade_et(self, kitap):
        for odunc in self.odunc_listesi:
            if odunc[0].kitap_id == kitap.kitap_id:
                kitap.durum_guncelle(True)
                self.odunc_listesi.remove(odunc)
                print("Kitap başarıyla iade edildi.")
                return
        print("Belirtilen kitap ödünçte değil.")

    def odunc_bilgisi(self):
        if not self.odunc_listesi:
            print("Ödünç kitap bulunmamaktadır.")
        else:
            print("Ödünç Kitap Bilgisi:")
            for odunc in self.odunc_listesi:
                print(f"Kitap ID: {odunc[0].kitap_id}, Adı: {odunc[0].ad}, Yazarı: {odunc[0].yazar}, Ödünç Alan: {odunc[1].ad} {odunc[1].soyad}, Tarih: {odunc[2]}")


# Örnek veri oluşturma
kitaplar = [
    Kitap(1, "Harry Potter ve Felsefe Taşı", "J.K. Rowling"),
    Kitap(2, "1984", "George Orwell"),
    Kitap(3, "Sefiller", "Victor Hugo"),
    Kitap(4, "Suç ve Ceza", "Fyodor Dostoyevski"),
    Kitap(5, "Yabancı", "Albert Camus")
]

uyeler = [
    Uye("Ahmet", "Yılmaz", 101),
    Uye("Ayşe", "Kaya", 102),
    Uye("Mehmet", "Öztürk", 103)
]

odunc_islemi = Odunc()
odunc_islemi.odunc_al(kitaplar[0], uyeler[0], "2024-04-01")
odunc_islemi.odunc_al(kitaplar[1], uyeler[1], "2024-04-02")
odunc_islemi.odunc_al(kitaplar[2], uyeler[2], "2024-04-03")
```

```python
import tkinter as tk
from tkinter import messagebox

class KutuphaneSistemiGUI:
    def __init__(self, root):
        self.root = root
        self.root.title("Kütüphane Yönetim Sistemi")

        self.kitaplar = [
            Kitap(1, "Harry Potter ve Felsefe Taşı", "J.K. Rowling"),
            Kitap(2, "1984", "George Orwell"),
            Kitap(3, "Sefiller", "Victor Hugo"),
            Kitap(4, "Suç ve Ceza", "Fyodor Dostoyevski"),
            Kitap(5, "Yabancı", "Albert Camus")
        ]

        self.uyeler = [
            Uye("Ahmet", "Yılmaz", 101),
            Uye("Ayşe", "Kaya", 102),
            Uye("Mehmet", "Öztürk", 103)
        ]

        self.odunc_islemi = Odunc()

        self.kitap_label = tk.Label(root, text="Kitap Seçimi:")
        self.kitap_label.grid(row=0, column=0)
        self.kitap_var = tk.StringVar(root)
        self.kitap_var.set(self.kitaplar[0].ad)
        self.kitap_optionmenu = tk.OptionMenu(root, self.kitap_var, *[k.ad for k in self.kitaplar])
        self.kitap_optionmenu.grid(row=0, column=1)

        self.uye_label = tk.Label(root, text="Üye Seçimi:")
        self.uye_label.grid(row=1, column=0)
        self.uye_var = tk.StringVar(root)
        self.uye_var.set(self.uyeler[0].ad)
        self.uye_optionmenu = tk.OptionMenu(root, self.uye_var, *[u.ad for u in self.uyeler])
        self.uye_optionmenu.grid(row=1, column=1)

        self.tarih_label = tk.Label(root, text="Tarih:")
        self.tarih_label.grid(row=2, column=0)
        self.tarih_entry = tk.Entry(root)
        self.tarih_entry.grid(row=2, column=1)

        self.odunc_button = tk.Button(root, text="Ödünç Al", command=self.odunc_al)
        self.odunc_button.grid(row=3, column=0, columnspan=2)

        self.iade_button = tk.Button(root, text="İade Et", command=self.iade_et)
        self.iade_button.grid(row=4, column=0, columnspan=2)

        self.bilgi_button = tk.Button(root, text="Ödünç Bilgisi", command=self.odunc_bilgisi)
        self.bilgi_button.grid(row=5, column=0, columnspan=2)

    def odunc_al(self):
        kitap_secimi = self.kitap_var.get()
        uye_secimi = self.uye_var.get()
        tarih = self.tarih_entry.get()

        secilen_kitap = next((k for k in self.kitaplar if k.ad == kitap_secimi), None)
        secilen_uye = next((u for u in self.uyeler if u.ad == uye_secimi), None)

        if secilen_kitap and secilen_uye:
            self.odunc_islemi.odunc_al(secilen_kitap, secilen_uye, tarih)
            messagebox.showinfo("Bilgi", "Kitap ödünç alındı.")
        else:
            messagebox.showerror("Hata", "Seçilen kitap veya üye bulunamadı.")

    def iade_et(self):
        kitap_secimi = self.kitap_var.get()
        secilen_kitap = next((k for k in self.kitaplar if k.ad == kitap_secimi), None)
        if secilen_kitap:
            self.odunc_islemi.iade_et(secilen_kitap)
        else:
            messagebox.showerror("Hata", "Seçilen kitap bulunamadı.")

    def odunc_bilgisi(self):
        self.odunc_islemi.odunc_bilgisi()


root = tk.Tk()
app = KutuphaneSistemiGUI(root)
root.mainloop()
```
