```python
import tkinter as tk
from random import randint

class Etkinlik:
    def __init__(self, adi, tarih, mekan):
        self.adi = adi
        self.tarih = tarih
        self.mekan = mekan

    def etkinlik_ekle(self):
        # Etkinlik ekleme işlemi burada yapılacak
        pass

    def bilet_sat(self, bilet):
        # Bilet satışı işlemi burada yapılacak
        pass

    def bilet_al(self):
        # Bilet alma işlemi burada yapılacak
        pass


class Bilet:
    def __init__(self, etkinlik):
        self.numara = self.rasgele_bilet_numarasi()
        self.etkinlik = etkinlik

    def rasgele_bilet_numarasi(self):
        return str(randint(100, 999))

class Kullanici:
    def __init__(self, adi, soyadi, email):
        self.adi = adi
        self.soyadi = soyadi
        self.email = email
        self.biletler = []

    def bilet_al(self, bilet):
        self.biletler.append(bilet)

    def biletlerimi_goruntule(self):
        for bilet in self.biletler:
            print("Bilet Numarası:", bilet.numara)
            print("Etkinlik:", bilet.etkinlik.adi)
            print("Tarih:", bilet.etkinlik.tarih)
            print("Mekan:", bilet.etkinlik.mekan)
            print()

class EtkinlikVeBiletUygulamasi:
    def __init__(self, ana_pencere):
        self.ana_pencere = ana_pencere
        ana_pencere.title("Etkinlik ve Bilet Satış Platformu")

        self.etkinlikler = [
            Etkinlik("Konser", "25.06.2024", "İstanbul Kongre Merkezi"),
            Etkinlik("Tiyatro Gösterisi", "30.06.2024", "Şehir Tiyatrosu"),
            Etkinlik("Şampiyonlar Ligi Maçı", "05.07.2024", "Olimpiyat Stadyumu"),
            # Diğer etkinlikler
        ]

        self.kullanici = Kullanici("Ahmet", "Yılmaz", "ahmet@example.com")

        self.etkinlikleri_goster_label = tk.Label(ana_pencere, text="Etkinlikleri Göster", font=("Arial", 14))
        self.etkinlikleri_goster_label.pack()

        self.etkinlik_listesi = tk.Listbox(ana_pencere, height=10, width=50)
        for etkinlik in self.etkinlikler:
            self.etkinlik_listesi.insert(tk.END, etkinlik.adi + " - " + etkinlik.tarih + " - " + etkinlik.mekan)
        self.etkinlik_listesi.pack()

        self.bilet_sat_button = tk.Button(ana_pencere, text="Bilet Satın Al", command=self.bilet_sat)
        self.bilet_sat_button.pack()

        self.biletlerim_button = tk.Button(ana_pencere, text="Biletlerimi Görüntüle", command=self.biletlerimi_goruntule)
        self.biletlerim_button.pack()


    def bilet_sat(self):
        secilen_index = self.etkinlik_listesi.curselection()
        if secilen_index:
            secilen_etkinlik = self.etkinlikler[secilen_index[0]]
            bilet = Bilet(secilen_etkinlik)
            self.kullanici.bilet_al(bilet)
            print("Bilet Satın Alındı:")
            print("Bilet Numarası:", bilet.numara)
            print("Etkinlik:", secilen_etkinlik.adi)
            print("Tarih:", secilen_etkinlik.tarih)
            print("Mekan:", secilen_etkinlik.mekan)
        else:
            print("Lütfen bir etkinlik seçin.")

    def biletlerimi_goruntule(self):
        self.kullanici.biletlerimi_goruntule()

root = tk.Tk()
uygulama = EtkinlikVeBiletUygulamasi(root)
root.mainloop()
```
