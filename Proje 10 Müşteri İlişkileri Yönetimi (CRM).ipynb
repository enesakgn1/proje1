```python
import tkinter as tk

class Musteri:
    def __init__(self, ad, telefon):
        self.ad = ad
        self.telefon = telefon

class Satış:
    def __init__(self, no, urunler):
        self.no = no
        self.urunler = urunler

class Destek:
    def __init__(self, no, detaylar):
        self.no = no
        self.detaylar = detaylar

musteriler = [
    Musteri("Emre", "54236789"),
    Musteri("Mehmet", "587654321"),
    Musteri("Uras", "587654321")
]

satislar = [
    Satış(1, ["Gitar"]),
    Satış(2, ["Piyano"]),
    Satış(3, ["Ney"])
]

destek_talepleri = [
    Destek(1, "İade alış yapmak istiyorum"),
    Destek(2, "Kargom nerede?"),
    Destek(3, "Başka bir talep oluşturcam")
]

def musteri_ekle():
    ad = musteri_ad.get()
    telefon = musteri_telefon.get()
    if ad and telefon:
        musteri = Musteri(ad, telefon)
        musteriler.append(musteri)
        print("Müşteri eklendi:", ad)
        musteri_ad.delete(0, tk.END)
        musteri_telefon.delete(0, tk.END)
    else:
        print("Lütfen müşteri adı ve telefonunu girin.")

def satış_ekle():
    no = satış_no.get()
    urunler = satış_urunler.get()
    if no and urunler:
        satış = Satış(no, urunler.split(", "))
        satislar.append(satış)
        print("Satış eklendi:", no)
        satış_no.delete(0, tk.END)
        satış_urunler.delete(0, tk.END)
    else:
        print("Lütfen satış numarası ve ürünleri girin.")

def destek_talebi_oluştur():
    no = destek_no.get()
    detaylar = destek_detaylar.get()
    if no and detaylar:
        destek_talebi = Destek(no, detaylar)
        destek_talepleri.append(destek_talebi)
        print("Destek talebi oluşturuldu:", no)
        destek_no.delete(0, tk.END)
        destek_detaylar.delete(0, tk.END)
    else:
        print("Lütfen destek talep numarası ve detaylarını girin.")

pencere = tk.Tk()
pencere.title("Müşteri İlişkileri Yönetimi (CRM)")

musteri_frame = tk.Frame(pencere)
musteri_frame.pack(pady=10)

tk.Label(musteri_frame, text="Müşteri Adı:").grid(row=0, column=0, padx=5, pady=5)
musteri_ad = tk.Entry(musteri_frame)
musteri_ad.grid(row=0, column=1, padx=5, pady=5)

tk.Label(musteri_frame, text="Telefon:").grid(row=1, column=0, padx=5, pady=5)
musteri_telefon = tk.Entry(musteri_frame)
musteri_telefon.grid(row=1, column=1, padx=5, pady=5)

musteri_ekle_button = tk.Button(musteri_frame, text="Müşteri Ekle", command=musteri_ekle)
musteri_ekle_button.grid(row=2, columnspan=2, pady=5)

satış_frame = tk.Frame(pencere)
satış_frame.pack(pady=10)

tk.Label(satış_frame, text="Satış No:").grid(row=0, column=0, padx=5, pady=5)
satış_no = tk.Entry(satış_frame)
satış_no.grid(row=0, column=1, padx=5, pady=5)

tk.Label(satış_frame, text="Ürünler:").grid(row=1, column=0, padx=5, pady=5)
satış_urunler = tk.Entry(satış_frame)
satış_urunler.grid(row=1, column=1, padx=5, pady=5)

satış_ekle_button = tk.Button(satış_frame, text="Satış Ekle", command=satış_ekle)
satış_ekle_button.grid(row=2, columnspan=2, pady=5)

destek_frame = tk.Frame(pencere)
destek_frame.pack(pady=10)

tk.Label(destek_frame, text="Destek Talep No:").grid(row=0, column=0, padx=5, pady=5)
destek_no = tk.Entry(destek_frame)
destek_no.grid(row=0, column=1, padx=5, pady=5)

tk.Label(destek_frame, text="Detaylar:").grid(row=1, column=0, padx=5, pady=5)
destek_detaylar = tk.Entry(destek_frame)
destek_detaylar.grid(row=1, column=1, padx=5, pady=5)

destek_talebi_oluştur_button = tk.Button(destek_frame, text="Destek Talebi Oluştur", command=destek_talebi_oluştur)
destek_talebi_oluştur_button.grid(row=2, columnspan=2, pady=5)

pencere.mainloop()
```
