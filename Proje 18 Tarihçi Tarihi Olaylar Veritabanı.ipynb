```python
import tkinter as tk
from tkinter import ttk

class Olay:
    def __init__(self, adı, tarihi, açıklama):
        self.adı = adı
        self.tarihi = tarihi
        self.açıklama = açıklama

class Şahsiyet:
    def __init__(self, adı, dönemler):
        self.adı = adı
        self.dönemler = dönemler

class Dönem:
    def __init__(self, adı, başlangıç_tarihi, bitiş_tarihi):
        self.adı = adı
        self.başlangıç_tarihi = başlangıç_tarihi
        self.bitiş_tarihi = bitiş_tarihi

class TarihiOlaylarUygulaması:
    def __init__(self, root):
        self.root = root
        self.root.title("Tarihi Olaylar Veritabanı")

        self.notebook = ttk.Notebook(root)
        self.notebook.pack(fill='both', expand=True)

        self.olaylar_frame = ttk.Frame(self.notebook)
        self.şahsiyetler_frame = ttk.Frame(self.notebook)
        self.dönemler_frame = ttk.Frame(self.notebook)

        self.notebook.add(self.olaylar_frame, text="Olaylar")
        self.notebook.add(self.şahsiyetler_frame, text="Şahsiyetler")
        self.notebook.add(self.dönemler_frame, text="Dönemler")

        self.olaylar_tree = ttk.Treeview(self.olaylar_frame, columns=("Tarih", "Açıklama"))
        self.olaylar_tree.heading("#0", text="Olay Adı")
        self.olaylar_tree.heading("#1", text="Tarih")
        self.olaylar_tree.heading("#2", text="Açıklama")
        self.olaylar_tree.pack(fill="both", expand=True)

        self.şahsiyetler_tree = ttk.Treeview(self.şahsiyetler_frame, columns=("Dönemler",))
        self.şahsiyetler_tree.heading("#0", text="Şahsiyet Adı")
        self.şahsiyetler_tree.heading("#1", text="Dönemler")
        self.şahsiyetler_tree.pack(fill="both", expand=True)

        self.dönemler_tree = ttk.Treeview(self.dönemler_frame, columns=("Başlangıç", "Bitiş"))
        self.dönemler_tree.heading("#0", text="Dönem Adı")
        self.dönemler_tree.heading("#1", text="Başlangıç")
        self.dönemler_tree.heading("#2", text="Bitiş")
        self.dönemler_tree.pack(fill="both", expand=True)

        self.olaylar = [
            Olay("Fransız Devrimi", "1789", "Fransa'da monarşinin devrilmesi ve cumhuriyetin ilan edilmesi."),
            Olay("Amerikan Bağımsızlık Bildirgesi", "1776", "Amerika Birleşik Devletleri'nin bağımsızlık ilanı."),
            Olay("İlk Dünya Savaşı", "1914-1918", "Birçok uluslararası gücün katıldığı büyük bir savaş."),
            Olay("İlk İnsanın Ay'a Ayak Basması", "1969", "Neil Armstrong'un Ay'a ayak basması."),
            Olay("Berlin Duvarı'nın Yıkılması", "1989", "Doğu ve Batı Almanya arasındaki sınırın kaldırılması.")
        ]

        self.şahsiyetler = [
            Şahsiyet("Mustafa Kemal Atatürk", ["Kurtuluş Savaşı", "Cumhuriyet Dönemi"]),
            Şahsiyet("Napolyon Bonaparte", ["Fransız Devrimi", "Napolyon Savaşları"]),
            Şahsiyet("Albert Einstein", ["20. yüzyıl"]),
            Şahsiyet("George Washington", ["Amerikan Bağımsızlık Savaşı", "Amerikan Başkanlık Dönemi"]),
            Şahsiyet("Mao Zedong", ["Çin Devrimi", "Çin Cumhuriyeti"])
        ]

        self.dönemler = [
            Dönem("Rönesans", "14. yüzyıl", "17. yüzyıl"),
            Dönem("Osmanlı İmparatorluğu", "1299", "1922"),
            Dönem("İlkçağ", "MÖ 3000", "MS 476"),
            Dönem("Ortaçağ", "476", "1453"),
            Dönem("Yeniçağ", "1453", "1789")
        ]

        self.güncelle()

    def güncelle(self):
        self.olaylar_tree.delete(*self.olaylar_tree.get_children())
        for olay in self.olaylar:
            self.olaylar_tree.insert("", "end", text=olay.adı, values=(olay.tarihi, olay.açıklama))

        self.şahsiyetler_tree.delete(*self.şahsiyetler_tree.get_children())
        for şahsiyet in self.şahsiyetler:
            self.şahsiyetler_tree.insert("", "end", text=şahsiyet.adı, values=(" - ".join(şahsiyet.dönemler),))

        self.dönemler_tree.delete(*self.dönemler_tree.get_children())
        for dönem in self.dönemler:
            self.dönemler_tree.insert("", "end", text=dönem.adı, values=(dönem.başlangıç_tarihi, dönem.bitiş_tarihi))


root = tk.Tk()
app = TarihiOlaylarUygulaması(root)
root.mainloop()
```
