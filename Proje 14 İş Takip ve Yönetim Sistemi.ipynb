```python
class Çalışan:
    def __init__(self, adı):
        self.adı = adı

class Görev:
    def __init__(self, adı, proje, çalışan):
        self.adı = adı
        self.proje = proje
        self.çalışan = çalışan

class Proje:
    def __init__(self, adı, başlangıç_tarihi, bitiş_tarihi):
        self.adı = adı
        self.başlangıç_tarihi = başlangıç_tarihi
        self.bitiş_tarihi = bitiş_tarihi
        self.görevler = []

    def görev_ata(self, görev):
        self.görevler.append(görev)
```

```python
# Çalışanları oluştur
çalışan1 = Çalışan("Ahmet")
çalışan2 = Çalışan("Mehmet")
çalışan3 = Çalışan("Ayşe")

# Projeleri oluştur
proje1 = Proje("Web Sitesi Yenileme", "2024-05-01", "2024-06-30")
proje2 = Proje("Mobil Uygulama Geliştirme", "2024-07-01", "2024-08-31")

# Görevleri oluştur
görev1 = Görev("Tasarım Hazırlığı", proje1, çalışan1)
görev2 = Görev("Frontend Geliştirme", proje1, çalışan2)
görev3 = Görev("Backend Geliştirme", proje2, çalışan3)

# Projelere görevleri ata
proje1.görev_ata(görev1)
proje1.görev_ata(görev2)
proje2.görev_ata(görev3)
```

```python
import tkinter as tk

class İşTakipArayüzü:
    def __init__(self, root):
        self.root = root
        self.root.title("İş Takip ve Yönetim Sistemi")

        self.proje_label = tk.Label(root, text="Proje Seçiniz:")
        self.proje_label.pack()

        self.proje_listbox = tk.Listbox(root, width=40, height=5, activestyle='dotbox')
        self.proje_listbox.pack()

        self.görev_label = tk.Label(root, text="Görev Seçiniz:")
        self.görev_label.pack()

        self.görev_listbox = tk.Listbox(root, width=40, height=5, activestyle='dotbox')
        self.görev_listbox.pack()

        self.çalışan_label = tk.Label(root, text="Çalışan Seçiniz:")
        self.çalışan_label.pack()

        self.çalışan_listbox = tk.Listbox(root, width=40, height=5, activestyle='dotbox')
        self.çalışan_listbox.pack()

        self.göster_button = tk.Button(root, text="Görevleri Göster", command=self.görevleri_göster)
        self.göster_button.pack()

    def görevleri_göster(self):
        seçili_proje = self.proje_listbox.get(tk.ACTIVE)
        seçili_görev = self.görev_listbox.get(tk.ACTIVE)
        seçili_çalışan = self.çalışan_listbox.get(tk.ACTIVE)

        görev_bilgisi = f"Proje: {seçili_proje}\nGörev: {seçili_görev}\nÇalışan: {seçili_çalışan}"

        self.görev_detay_label = tk.Label(root, text=görev_bilgisi, font=("Arial", 12))
        self.görev_detay_label.pack()

root = tk.Tk()
app = İşTakipArayüzü(root)

# Projeleri listbox'a ekle
for proje in [proje1, proje2]:
    app.proje_listbox.insert(tk.END, proje.adı)

# Görevleri listbox'a ekle
for görev in proje1.görevler:
    app.görev_listbox.insert(tk.END, görev.adı)

# Çalışanları listbox'a ekle
for çalışan in [çalışan1, çalışan2, çalışan3]:
    app.çalışan_listbox.insert(tk.END, çalışan.adı)

root.mainloop()
```
