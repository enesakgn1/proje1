```python
import tkinter as tk
from tkinter import ttk

class Oyun:
    def __init__(self, adı, türü, platformu):
        self.adı = adı
        self.türü = türü
        self.platformu = platformu

class Öneri:
    def __init__(self, oyun_adı, içerik):
        self.oyun_adı = oyun_adı
        self.içerik = içerik

class OyunKoleksiyonuUygulaması:
    def __init__(self, root):
        self.root = root
        self.root.title("Video Oyun Koleksiyonu Yönetimi")

        self.oyunlar = [
            Oyun("The Witcher 3", "RPG", "PC"),
            Oyun("Super Mario Odyssey", "Platform", "Nintendo Switch"),
            Oyun("The Last of Us Part II", "Aksiyon", "PlayStation 4"),
            Oyun("The Legend of Zelda: Breath of the Wild", "Aksiyon", "Nintendo Switch"),
            Oyun("Grand Theft Auto V", "Açık Dünya", "PC"),
            Oyun("FIFA 21", "Spor", "PlayStation 4"),
            Oyun("Minecraft", "Sandbox", "PC"),
            Oyun("Call of Duty: Warzone", "FPS", "PC"),
            Oyun("Animal Crossing: New Horizons", "Simülasyon", "Nintendo Switch"),
            Oyun("Among Us", "Party", "PC")
        ]

        self.öneriler = [
            Öneri("The Witcher 3", "Büyüleyici bir dünyada epik bir maceraya atılın."),
            Öneri("Super Mario Odyssey", "Mario ile eğlenceli platform maceralarına dalın."),
            Öneri("The Last of Us Part II", "Heyecan dolu bir hikaye ve nefes kesen aksiyon."),
            Öneri("Grand Theft Auto V", "Özgürce keşfedin ve suç dünyasında yolunuzu bulun."),
            Öneri("FIFA 21", "Gerçekçi futbol deneyimi yaşayın ve şampiyon olun."),
            Öneri("Minecraft", "Sonsuz bir yaratıcılıkla kendi dünyanızı inşa edin."),
            Öneri("Call of Duty: Warzone", "Yoğun çok oyunculu savaşlara katılın."),
            Öneri("Animal Crossing: New Horizons", "Huzurlu bir adada yaşamın tadını çıkarın."),
            Öneri("Among Us", "Arkadaşlarınızla eğlenceli bir dedektiflik oyunu oynayın.")
        ]

        self.koleksiyon_label = tk.Label(self.root, text="Oyunlar")
        self.koleksiyon_label.pack()

        self.koleksiyon_tree = ttk.Treeview(self.root, columns=("Tür", "Platform"))
        self.koleksiyon_tree.heading("#0", text="Oyun Adı")
        self.koleksiyon_tree.heading("#1", text="Tür")
        self.koleksiyon_tree.heading("#2", text="Platform")
        self.koleksiyon_tree.pack(fill="both", expand=True)

        for oyun in self.oyunlar:
            self.koleksiyon_tree.insert("", "end", text=oyun.adı, values=(oyun.türü, oyun.platformu))

        self.favori_label = tk.Label(self.root, text="Favori Oyunlar")
        self.favori_label.pack()

        self.favori_tree = ttk.Treeview(self.root, columns=("Tür", "Platform"))
        self.favori_tree.heading("#0", text="Oyun Adı")
        self.favori_tree.heading("#1", text="Tür")
        self.favori_tree.heading("#2", text="Platform")
        self.favori_tree.pack(fill="both", expand=True)

        self.favori_ekle_button = tk.Button(self.root, text="Favorilere Ekle", command=self.favori_ekle)
        self.favori_ekle_button.pack()

        self.öneri_al_button = tk.Button(self.root, text="Önerileri Al", command=self.öneri_al)
        self.öneri_al_button.pack()

        self.öneri_label = tk.Label(self.root, text="")
        self.öneri_label.pack()

        self.öneri_tree = ttk.Treeview(self.root, columns=("İçerik",))
        self.öneri_tree.heading("#0", text="Oyun Adı")
        self.öneri_tree.pack(fill="both", expand=True)

    def favori_ekle(self):
        selection = self.koleksiyon_tree.selection()
        if selection:
            item = self.koleksiyon_tree.item(selection, "text")
            values = self.koleksiyon_tree.item(selection, "values")
            self.favori_tree.insert("", "end", text=item, values=(values[0], values[1]))

    def öneri_al(self):
        self.öneri_label.config(text="Önerilen oyunlar:")
        self.öneri_tree.delete(*self.öneri_tree.get_children())
        for öneri in self.öneriler:
            self.öneri_tree.insert("", "end", text=öneri.oyun_adı, values=(öneri.içerik,))


root = tk.Tk()
app = OyunKoleksiyonuUygulaması(root)
root.mainloop()
```
