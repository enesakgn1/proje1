```python
class Seyahat:
    def __init__(self):
        self.rotalar = []

    def rota_ekle(self, rota):
        self.rotalar.append(rota)

    def plan_düzenle(self, eski_rota, yeni_rota):
        if eski_rota in self.rotalar:
            index = self.rotalar.index(eski_rota)
            self.rotalar[index] = yeni_rota
            return True
        return False

class Konaklama:
    def __init__(self, ad, fiyat):
        self.ad = ad
        self.fiyat = fiyat

class Rota:
    def __init__(self, rota_detayları):
        self.rota_detayları = rota_detayları
```

```python
konaklama_örnekleri = [
    Konaklama("Otel 1", 100),
    Konaklama("Otel 2", 150),
    Konaklama("Otel 3", 200),
    Konaklama("Pansiyon 1", 80),
    Konaklama("Pansiyon 2", 120),
    Konaklama("Tatil Köyü 1", 180),
    Konaklama("Tatil Köyü 2", 220),
    Konaklama("Hostel 1", 50),
    Konaklama("Hostel 2", 70),
    Konaklama("Ev Kiralama 1", 250)
]

rota_örnekleri = [
    Rota("Türkiye"),
    Rota("İtalya"),
    Rota("Fransa"),
    Rota("İspanya"),
    Rota("Yunanistan"),
    Rota("Almanya"),
    Rota("Hollanda"),
    Rota("İngiltere"),
    Rota("Portekiz"),
    Rota("Amerika")
]

```

```python
import tkinter as tk
from tkinter import messagebox

class SeyahatArayuzu:
    def __init__(self, root, seyahat):
        self.root = root
        self.seyahat = seyahat
        self.root.title("Seyahat Planlama Uygulaması")

        self.rota_label = tk.Label(root, text="Seyahat Rotası:")
        self.rota_label.grid(row=0, column=0, padx=10, pady=10)

        self.rota_liste = tk.Listbox(root)
        self.rota_liste.grid(row=0, column=1, padx=10, pady=10)

        self.rota_ekle_button = tk.Button(root, text="Rota Ekle", command=self.rota_ekle)
        self.rota_ekle_button.grid(row=0, column=2, padx=10, pady=10)

        self.rota_liste.bind("<<ListboxSelect>>", self.konaklama_goster)

        self.konaklama_label = tk.Label(root, text="Konaklama Seçenekleri:")
        self.konaklama_label.grid(row=1, column=0, padx=10, pady=10)

        self.konaklama_liste = tk.Listbox(root)
        self.konaklama_liste.grid(row=1, column=1, padx=10, pady=10)

        self.konaklama_liste.bind("<<ListboxSelect>>", self.konaklama_sec)

        self.örnekler_ekle()

    def rota_ekle(self):
        rota = f"Rota {len(self.seyahat.rotalar) + 1}"
        self.seyahat.rota_ekle(rota)
        self.rota_liste.insert(tk.END, rota)

    def konaklama_goster(self, event):
        secili_rota = self.rota_liste.curselection()
        if secili_rota:
            secili_rota = self.rota_liste.get(secili_rota)
            self.konaklama_liste.delete(0, tk.END)
            for konaklama in konaklama_örnekleri:
                self.konaklama_liste.insert(tk.END, f"{secili_rota} - {konaklama.ad} - {konaklama.fiyat} €")

    def konaklama_sec(self, event):
        secili_konaklama = self.konaklama_liste.curselection()
        if secili_konaklama:
            secili_konaklama = self.konaklama_liste.get(secili_konaklama)
            messagebox.showinfo("Seçilen Konaklama", f"Seçilen konaklama: {secili_konaklama}")

    def örnekler_ekle(self):
        for rota in rota_örnekleri:
            self.rota_liste.insert(tk.END, rota.rota_detayları)


if __name__ == "__main__":
    root = tk.Tk()
    seyahat = Seyahat()
    uygulama = SeyahatArayuzu(root, seyahat)
    root.mainloop()
```
