```python
class Etkinlik:
    def __init__(self, etkinlik_id, ad, tarih, yer):
        self.etkinlik_id = etkinlik_id
        self.ad = ad
        self.tarih = tarih
        self.yer = yer


class Katilimci:
    def __init__(self, ad, soyad, email):
        self.ad = ad
        self.soyad = soyad
        self.email = email


class Bilet:
    def __init__(self, bilet_id, etkinlik, katilimci):
        self.bilet_id = bilet_id
        self.etkinlik = etkinlik
        self.katilimci = katilimci


etkinlikler = [
    Etkinlik(1, "Konser", "2024-05-15", "Stadyum"),
    Etkinlik(2, "Sergi", "2024-06-20", "Sanat Galerisi"),
    Etkinlik(3, "Konferans", "2024-07-10", "Kongre Merkezi"),
    Etkinlik(4, "Film Gösterimi", "2024-08-05", "Sinema")
]

katilimcilar = [
    Katilimci("Ahmet", "Yılmaz", "ahmet@example.com"),
    Katilimci("Ayşe", "Kaya", "ayse@example.com"),
    Katilimci("Mehmet", "Öztürk", "mehmet@example.com")
]

biletler = [
    Bilet(1, etkinlikler[0], katilimcilar[0]),
    Bilet(2, etkinlikler[1], katilimcilar[1]),
    Bilet(3, etkinlikler[2], katilimcilar[2])
]
```

```python
import tkinter as tk
from tkinter import messagebox

class EtkinlikYonetimSistemiGUI:
    def __init__(self, root):
        self.root = root
        self.root.title("Etkinlik Yönetim Sistemi")

        self.etkinlikler = [
            Etkinlik(1, "Konser", "2024-05-15", "Stadyum"),
            Etkinlik(2, "Sergi", "2024-06-20", "Sanat Galerisi"),
            Etkinlik(3, "Konferans", "2024-07-10", "Kongre Merkezi"),
            Etkinlik(4, "Film Gösterimi", "2024-08-05", "Sinema")
        ]

        self.katilimcilar = [
            Katilimci("Ahmet", "Yılmaz", "ahmet@example.com"),
            Katilimci("Ayşe", "Kaya", "ayse@example.com"),
            Katilimci("Mehmet", "Öztürk", "mehmet@example.com")
        ]

        self.biletler = [
            Bilet(1, self.etkinlikler[0], self.katilimcilar[0]),
            Bilet(2, self.etkinlikler[1], self.katilimcilar[1]),
            Bilet(3, self.etkinlikler[2], self.katilimcilar[2])
        ]

        self.etkinlik_label = tk.Label(root, text="Etkinlik Seçimi:")
        self.etkinlik_label.grid(row=0, column=0)
        self.etkinlik_var = tk.StringVar(root)
        self.etkinlik_var.set(self.etkinlikler[0].ad)
        self.etkinlik_optionmenu = tk.OptionMenu(root, self.etkinlik_var, *[e.ad for e in self.etkinlikler])
        self.etkinlik_optionmenu.grid(row=0, column=1)

        self.katilimci_label = tk.Label(root, text="Katılımcı Seçimi:")
        self.katilimci_label.grid(row=1, column=0)
        self.katilimci_var = tk.StringVar(root)
        self.katilimci_var.set(self.katilimcilar[0].ad)
        self.katilimci_optionmenu = tk.OptionMenu(root, self.katilimci_var, *[k.ad for k in self.katilimcilar])
        self.katilimci_optionmenu.grid(row=1, column=1)

        self.bilet_al_button = tk.Button(root, text="Bilet Al", command=self.bilet_al)
        self.bilet_al_button.grid(row=2, column=0, columnspan=2)

        self.bilet_bilgisi_button = tk.Button(root, text="Bilet Bilgisi", command=self.bilet_bilgisi)
        self.bilet_bilgisi_button.grid(row=3, column=0, columnspan=2)

    def bilet_al(self):
        etkinlik_secimi = self.etkinlik_var.get()
        katilimci_secimi = self.katilimci_var.get()

        secilen_etkinlik = next((e for e in self.etkinlikler if e.ad == etkinlik_secimi), None)
        secilen_katilimci = next((k for k in self.katilimcilar if k.ad == katilimci_secimi), None)

        if secilen_etkinlik and secilen_katilimci:
            yeni_bilet_id = len(self.biletler) + 1
            yeni_bilet = Bilet(yeni_bilet_id, secilen_etkinlik, secilen_katilimci)
            self.biletler.append(yeni_bilet)
            messagebox.showinfo("Bilgi", "Bilet başarıyla alındı.")
        else:
            messagebox.showerror("Hata", "Seçilen etkinlik veya katılımcı bulunamadı.")

    def bilet_bilgisi(self):
        if not self.biletler:
            messagebox.showinfo("Bilgi", "Bilet bulunmamaktadır.")
        else:
            bilgi = "\n".join([f"Bilet ID: {b.bilet_id}, Etkinlik: {b.etkinlik.ad}, Katılımcı: {b.katilimci.ad}" for b in self.biletler])
            messagebox.showinfo("Bilet Bilgisi", bilgi)


root = tk.Tk()
app = EtkinlikYonetimSistemiGUI(root)
root.mainloop()
```
