```python
import tkinter as tk
from tkinter import messagebox

class Kurs:
    def __init__(self, kurs_id, ad, egitmen, icerik=None, ogrenciler=None):
        self.kurs_id = kurs_id
        self.ad = ad
        self.egitmen = egitmen
        self.icerik = icerik if icerik else []
        self.ogrenciler = ogrenciler if ogrenciler else []

    def kurs_olustur(self):
        print(f"{self.ad} kursu oluşturuldu.")

    def kaydol(self, ogrenci):
        self.ogrenciler.append(ogrenci)
        print(f"{ogrenci.isim}, {self.ad} kursuna kaydoldu.")

    def icerik_yukle(self, icerik):
        self.icerik.append(icerik)
        print(f"{icerik} içerik yüklendi.")

class Egitmen:
    def __init__(self, isim, uzmanlik_alani):
        self.isim = isim
        self.uzmanlik_alani = uzmanlik_alani

class Ogrenci:
    def __init__(self, isim, email, tc, telefon):
        self.isim = isim
        self.email = email
        self.tc = tc
        self.telefon = telefon

class EgitimPlatformuGUI:
    def __init__(self, root):
        self.root = root
        self.root.title("Online Eğitim Platformu")

        self.kurslar = [
            Kurs(1, "Python Programlama", egitmen1),
            Kurs(2, "Veri Bilimi", egitmen2),
            Kurs(3, "Web Geliştirme", egitmen3)
        ]

        self.kayit_label = tk.Label(root, text="Kurs Seçimi:")
        self.kayit_label.grid(row=0, column=0)
        self.kayit_var = tk.StringVar(root)
        self.kayit_var.set(self.kurslar[0].ad)
        self.kayit_optionmenu = tk.OptionMenu(root, self.kayit_var, *[k.ad for k in self.kurslar])
        self.kayit_optionmenu.grid(row=0, column=1)

        self.isim_label = tk.Label(root, text="İsim:")
        self.isim_label.grid(row=1, column=0)
        self.isim_entry = tk.Entry(root)
        self.isim_entry.grid(row=1, column=1)

        self.email_label = tk.Label(root, text="E-mail:")
        self.email_label.grid(row=2, column=0)
        self.email_entry = tk.Entry(root)
        self.email_entry.grid(row=2, column=1)

        self.tc_label = tk.Label(root, text="TC:")
        self.tc_label.grid(row=3, column=0)
        self.tc_entry = tk.Entry(root)
        self.tc_entry.grid(row=3, column=1)

        self.telefon_label = tk.Label(root, text="Telefon:")
        self.telefon_label.grid(row=4, column=0)
        self.telefon_entry = tk.Entry(root)
        self.telefon_entry.grid(row=4, column=1)

        self.kaydol_button = tk.Button(root, text="Kaydol", command=self.kaydol)
        self.kaydol_button.grid(row=5, column=0, columnspan=2)

    def validate_tc_telefon(self, tc, telefon):
        if len(tc) != 11 or not tc.isdigit():
            return False
        if len(telefon) != 11 or not telefon.isdigit():
            return False
        return True

    def kaydol(self):
        kurs_secimi = self.kayit_var.get()
        isim = self.isim_entry.get()
        email = self.email_entry.get()
        tc = self.tc_entry.get()
        telefon = self.telefon_entry.get()

        if not self.validate_tc_telefon(tc, telefon):
            messagebox.showerror("Hata", "TC ve telefon numarası 11 haneli olmalıdır ve sadece rakamlardan oluşmalıdır.")
            return

        secilen_kurs = next((k for k in self.kurslar if k.ad == kurs_secimi), None)

        if secilen_kurs:
            yeni_ogrenci = Ogrenci(isim, email, tc, telefon)
            secilen_kurs.kaydol(yeni_ogrenci)
            messagebox.showinfo("Bilgi", "Kaydınız başarıyla tamamlandı.")
        else:
            messagebox.showerror("Hata", "Seçilen kurs bulunamadı.")


egitmen1 = Egitmen("Ahmet Yılmaz", "Python Programlama")
egitmen2 = Egitmen("Ayşe Kaya", "Veri Bilimi")
egitmen3 = Egitmen("Mehmet Öztürk", "Web Geliştirme")

kurs1 = Kurs(1, "Python Programlama", egitmen1)
kurs2 = Kurs(2, "Veri Bilimi", egitmen2)
kurs3 = Kurs(3, "Web Geliştirme", egitmen3)

ogrenci1 = Ogrenci("Ali", "ali@example.com", "12345678901", "5555555555")
ogrenci2 = Ogrenci("Ayşe", "ayse@example.com", "23456789012", "5555555556")
ogrenci3 = Ogrenci("Fatma", "fatma@example.com", "34567890123", "5555555557")


root = tk.Tk()
app = EgitimPlatformuGUI(root)
root.mainloop()
```
