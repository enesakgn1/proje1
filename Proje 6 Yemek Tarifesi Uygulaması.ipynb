```python
class Tarif:
    def __init__(self, tarif_id, ad, malzemeler=None, tarif_icerigi=None, puan=0):
        self.tarif_id = tarif_id
        self.ad = ad
        self.malzemeler = malzemeler if malzemeler else []
        self.tarif_icerigi = tarif_icerigi if tarif_icerigi else ""
        self.puan = puan

    def tarif_ekle(self):
        print(f"{self.ad} tarifi eklendi.")

    def tarif_ara(self, anahtar_kelime):
        if anahtar_kelime.lower() in self.ad.lower() or anahtar_kelime.lower() in self.tarif_icerigi.lower():
            return True
        else:
            return False

    def tarif_degerlendir(self, puan):
        self.puan = (self.puan + puan) / 2
        print(f"{self.ad} tarifine {puan} puan verildi.")


class Malzeme:
    def __init__(self, ad, miktar):
        self.ad = ad
        self.miktar = miktar


class Kullanici:
    def __init__(self, kullanici_id, ad, sifre):
        self.kullanici_id = kullanici_id
        self.ad = ad
        self.sifre = sifre


# Örnek veri oluşturma
tarif1 = Tarif(1, "Omlet", [Malzeme("Yumurta", "2 adet"), Malzeme("Süt", "50 ml")], "Yumurtaları çırpın, süt ekleyin ve tavada pişirin.")
tarif2 = Tarif(2, "Makarna", [Malzeme("Makarna", "250 gram"), Malzeme("Sos", "150 ml")], "Makarnayı haşlayın, sosu ekleyip karıştırın.")
tarif3 = Tarif(3, "Salata", [Malzeme("Marul", "1 baş"), Malzeme("Domates", "2 adet")], "Malzemeleri doğrayın, karıştırın ve servis yapın.")
tarif4 = Tarif(4, "Köfte", [Malzeme("Kıyma", "500 gram"), Malzeme("Soğan", "1 adet")], "Kıymayı yoğurun, köfte şekli verin ve pişirin.")
tarif5 = Tarif(5, "Kahvaltı Tabağı", [Malzeme("Peynir", "100 gram"), Malzeme("Zeytin", "50 gram")], "Malzemeleri tabağa yerleştirin ve servis yapın.")
```

```python
import tkinter as tk
from tkinter import messagebox

class YemekTarifiGUI:
    def __init__(self, root):
        self.root = root
        self.root.title("Yemek Tarifi Uygulaması")

        self.tarifler = [
            Tarif(1, "Omlet", [Malzeme("Yumurta", "2 adet"), Malzeme("Süt", "50 ml")], "Yumurtaları çırpın, süt ekleyin ve tavada pişirin."),
            Tarif(2, "Makarna", [Malzeme("Makarna", "250 gram"), Malzeme("Sos", "150 ml")], "Makarnayı haşlayın, sosu ekleyip karıştırın."),
            Tarif(3, "Salata", [Malzeme("Marul", "1 baş"), Malzeme("Domates", "2 adet")], "Malzemeleri doğrayın, karıştırın ve servis yapın."),
            Tarif(4, "Köfte", [Malzeme("Kıyma", "500 gram"), Malzeme("Soğan", "1 adet")], "Kıymayı yoğurun, köfte şekli verin ve pişirin."),
            Tarif(5, "Kahvaltı Tabağı", [Malzeme("Peynir", "100 gram"), Malzeme("Zeytin", "50 gram")], "Malzemeleri tabağa yerleştirin ve servis yapın.")
        ]

        self.arama_label = tk.Label(root, text="Tarif Ara:")
        self.arama_label.grid(row=0, column=0)
        self.arama_entry = tk.Entry(root)
        self.arama_entry.grid(row=0, column=1)
        self.arama_button = tk.Button(root, text="Ara", command=self.tarif_ara)
        self.arama_button.grid(row=0, column=2)

        self.tarifler_listbox = tk.Listbox(root, width=50)
        self.tarifler_listbox.grid(row=1, column=0, columnspan=3)
        self.tarifler_listbox.bind('<<ListboxSelect>>', self.tarif_detaylari)

        self.detaylar_label = tk.Label(root, text="Tarif Detayları:")
        self.detaylar_label.grid(row=2, column=0, sticky='w')
        self.detaylar_text = tk.Text(root, width=50, height=5)
        self.detaylar_text.grid(row=3, column=0, columnspan=3)

        self.puan_label = tk.Label(root, text="Puan Ver:")
        self.puan_label.grid(row=4, column=0, sticky='w')
        self.puan_entry = tk.Entry(root, width=10)
        self.puan_entry.grid(row=4, column=1)
        self.puan_button = tk.Button(root, text="Puan Ver", command=self.tarif_puan_ver)
        self.puan_button.grid(row=4, column=2)

        self.tarifleri_goster()

    def tarifleri_goster(self):
        self.tarifler_listbox.delete(0, tk.END)
        for tarif in self.tarifler:
            self.tarifler_listbox.insert(tk.END, tarif.ad)

    def tarif_detaylari(self, event):
        selected_index = self.tarifler_listbox.curselection()
        if selected_index:
            selected_tarif = self.tarifler[selected_index[0]]
            self.detaylar_text.delete('1.0', tk.END)
            self.detaylar_text.insert(tk.END, f"Adı: {selected_tarif.ad}\n\nMalzemeler:\n")
            for malzeme in selected_tarif.malzemeler:
                self.detaylar_text.insert(tk.END, f"- {malzeme.ad}: {malzeme.miktar}\n")
            self.detaylar_text.insert(tk.END, f"\nTarif İçeriği:\n{selected_tarif.tarif_icerigi}")

    def tarif_ara(self):
        anahtar_kelime = self.arama_entry.get()
        bulunan_tarifler = [tarif for tarif in self.tarifler if tarif.tarif_ara(anahtar_kelime)]
        self.tarifler_listbox.delete(0, tk.END)
        for tarif in bulunan_tarifler:
            self.tarifler_listbox.insert(tk.END, tarif.ad)

    def tarif_puan_ver(self):
        selected_index = self.tarifler_listbox.curselection()
        if selected_index:
            selected_tarif = self.tarifler[selected_index[0]]
            puan = int(self.puan_entry.get())
            selected_tarif.tarif_degerlendir(puan)
            messagebox.showinfo("Bilgi", f"{selected_tarif.ad} tarifine {puan} puan verildi.")
            self.puan_entry.delete(0, tk.END)


root = tk.Tk()
app = YemekTarifiGUI(root)
root.mainloop()
```
