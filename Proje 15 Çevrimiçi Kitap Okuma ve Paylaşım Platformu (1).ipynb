```python
import tkinter as tk
from tkinter import ttk

class Kitap:
    def __init__(self, ad, yazar, yayınevi):
        self.ad = ad
        self.yazar = yazar
        self.yayınevi = yayınevi
        self.yorumlar = []

    def yorum_yap(self, yorum):
        self.yorumlar.append(yorum)

class Kullanıcı:
    def __init__(self, kullanıcı_adı, şifre):
        self.kullanıcı_adı = kullanıcı_adı
        self.şifre = şifre
        self.okunan_kitaplar = []

    def kitap_oku(self, kitap):
        self.okunan_kitaplar.append(kitap)

# Veri eklemesi
kitap1 = Kitap("Suç ve Ceza", "Fyodor Dostoyevski", "YKY")
kitap2 = Kitap("1984", "George Orwell", "Can Yayınları")
kitap3 = Kitap("Harry Potter ve Felsefe Taşı", "J.K. Rowling", "YKY")

kullanıcı1 = Kullanıcı("user1", "123456")
kullanıcı2 = Kullanıcı("user2", "password")
kullanıcı3 = Kullanıcı("user3", "abc123")

# Arayüz tasarımı
class KitapUygulaması:
    def __init__(self, pencere):
        self.pencere = pencere
        self.pencere.title("Çevrimiçi Kitap Okuma ve Paylaşım Platformu")

        self.kullanıcılar = [kullanıcı1, kullanıcı2, kullanıcı3]

        self.kullanıcı_combobox = tk.StringVar()
        self.kullanıcı_combobox.set(self.kullanıcılar[0].kullanıcı_adı)
        self.kullanıcı_combo = tk.OptionMenu(pencere, self.kullanıcı_combobox, *self.kullanıcı_adları())
        self.kullanıcı_combo.pack()

        self.kitap_combobox = tk.StringVar()
        self.kitap_combobox.set(kitap1.ad)
        self.kitap_combo = tk.OptionMenu(pencere, self.kitap_combobox, *self.kitap_adları())
        self.kitap_combo.pack()

        self.oku_button = tk.Button(pencere, text="Oku", command=self.kitap_oku)
        self.oku_button.pack()

        self.yorum_button = tk.Button(pencere, text="Yorum Yap", command=self.yorum_sekmeyi_aç)
        self.yorum_button.pack()

    def kullanıcı_adları(self):
        return [kullanıcı.kullanıcı_adı for kullanıcı in self.kullanıcılar]

    def kitap_adları(self):
        return [kitap.ad for kitap in [kitap1, kitap2, kitap3]]

    def kitap_oku(self):
        seçili_kitap_adı = self.kitap_combobox.get()
        seçili_kitap = None
        for kitap in [kitap1, kitap2, kitap3]:
            if kitap.ad == seçili_kitap_adı:
                seçili_kitap = kitap
                break

        seçili_kullanıcı_adı = self.kullanıcı_combobox.get()
        seçili_kullanıcı = None
        for kullanıcı in self.kullanıcılar:
            if kullanıcı.kullanıcı_adı == seçili_kullanıcı_adı:
                seçili_kullanıcı = kullanıcı
                break

        if seçili_kitap and seçili_kullanıcı:
            seçili_kullanıcı.kitap_oku(seçili_kitap)
            self.yazdır_okunan_kitaplar()

    def yorum_sekmeyi_aç(self):
        yorum_pencere = tk.Toplevel(self.pencere)
        yorum_pencere.title("Yorum Yap")

        seçili_kitap_adı = self.kitap_combobox.get()
        seçili_kitap = None
        for kitap in [kitap1, kitap2, kitap3]:
            if kitap.ad == seçili_kitap_adı:
                seçili_kitap = kitap
                break

        seçili_kullanıcı_adı = self.kullanıcı_combobox.get()
        seçili_kullanıcı = None
        for kullanıcı in self.kullanıcılar:
            if kullanıcı.kullanıcı_adı == seçili_kullanıcı_adı:
                seçili_kullanıcı = kullanıcı
                break

        yorum_label = tk.Label(yorum_pencere, text=f"Yorum Yap ({seçili_kitap.ad})")
        yorum_label.pack()

        yorum_text = tk.Text(yorum_pencere, height=5, width=50)
        yorum_text.pack()

        yorum_button = tk.Button(yorum_pencere, text="Yorum Yap", command=lambda: self.yorum_yap(seçili_kullanıcı, seçili_kitap, yorum_text.get("1.0", tk.END)))
        yorum_button.pack()

    def yorum_yap(self, kullanıcı, kitap, yorum_metni):
        if not yorum_metni.strip():
            return

        kitap.yorumlar.append(f"{kullanıcı.kullanıcı_adı}: {yorum_metni.strip()}")

        self.yorumları_görüntüle(kitap)

    def yorumları_görüntüle(self, kitap):
        yorum_pencere = tk.Toplevel(self.pencere)
        yorum_pencere.title(f"Yorumlar ({kitap.ad})")

        yorum_label = tk.Label(yorum_pencere, text=f"Yorumlar ({kitap.ad}):")
        yorum_label.pack()

        yorum_text = tk.Text(yorum_pencere, height=10, width=50)
        yorum_text.pack()

        for yorum in kitap.yorumlar:
            yorum_text.insert(tk.END, f"{yorum}\n")

        self.yazdır_okunan_kitaplar()

    def yazdır_okunan_kitaplar(self):
        okunan_pencere = tk.Toplevel(self.pencere)
        okunan_pencere.title("Okunan Kitaplar")

        okunan_label = tk.Label(okunan_pencere, text="Okunan Kitaplar:")
        okunan_label.pack()

        for kullanıcı in self.kullanıcılar:
            if kullanıcı.okunan_kitaplar:
                okunan_label = tk.Label(okunan_pencere, text=f"{kullanıcı.kullanıcı_adı} okudu")
                okunan_label.pack()
                for kitap in kullanıcı.okunan_kitaplar:
                    okunan_label = tk.Label(okunan_pencere, text=f"- {kitap.ad}")
                    okunan_label.pack()
            else:
                okunan_label = tk.Label(okunan_pencere, text=f"{kullanıcı.kullanıcı_adı} okumadı")
                okunan_label.pack()

if __name__ == "__main__":
    pencere = tk.Tk()
    uygulama = KitapUygulaması(pencere)
    pencere.mainloop()
```
